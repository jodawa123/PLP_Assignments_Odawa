```python
import pandas as pd
import numpy as np

df= pd.read_csv("/content/shopping_behavior_updated.csv")
df.head()

```

```python
df.info()

```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
groups=df.groupby("Age").mean(numeric_only=True)
print(groups)
```

```python
import matplotlib.pyplot as plt

# Group by Gender and sum total purchase amount
grouped = df.groupby("Gender")["Purchase Amount (USD)"].sum()

# Plot as a bar chart
plt.figure(figsize=(8,5))
grouped.plot(kind="bar", color=["skyblue", "pink"])
plt.title("Total Purchase Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Purchase Amount (USD)")
plt.show()
```

```python

age_trend = df.groupby("Age")["Purchase Amount (USD)"].mean()

plt.figure(figsize=(10,5))
plt.plot(age_trend.index, age_trend.values, marker='o', color='blue', label="Avg Purchase Amount")
plt.title("Average Purchase Amount by Age")
plt.xlabel("Age")
plt.ylabel("Average Purchase Amount (USD)")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.hist(df["Review Rating"], bins=10, color="purple", edgecolor="black")
plt.title("Distribution of Review Ratings")
plt.xlabel("Review Rating")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Purchase Amount (USD)"], alpha=0.6, c="green")
plt.title("Age vs Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```
